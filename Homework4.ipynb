```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import scipy
from scipy import stats
from scipy import signal
from scipy.stats import norm
from astropy.table import QTable, Table, Column
from astropy import units as u
import numpy as np

plt.rcParams["figure.figsize"] = (15, 10)
```

## Problem 1

```python
t = np.linspace(0,np.pi*100,np.int(1e5))
N = 10+np.sin(t)
bkgd = stats.norm.rvs(size=np.int(1e5))*np.sqrt(N)+N
```

Scatter plot of the first 1000 data points:

```python
x = t[:1000]
y = bkgd[:1000]
plt.plot(x, y, 'o')
plt.show()
```

I define a function that makes it so I can make a plot of any X contiguous points:

```python
def filter(x_array, y_array, start, stop):
    x = x_array[start:stop]
    y = y_array[start:stop]
    plt.plot(x, y, 'o')
```

Making a scatter plot of the data points between the 5000th and 6000th data points:

```python
filter(t, bkgd, 5000, 6000)
```

Plotting every 100th data point:

```python
x5 = t[::100]
y5 = bkgd[::100]
plt.plot(x5, y5, 'o')
plt.show()
```

## Problem 2

Making a sample histogram of the time vs background:

```python
plt.hist2d(t, bkgd, 200)
plt.colorbar()
plt.show()
```

The 2D histogram plots the density of occurrence of each combination of x and y in our input arrays. Each color on the plot corresponds to the number of occurences of each combination of x and y. In the plot above, the colors at the top of the colorbar indicate points on the 2D histogram corresponding the a larger number of occurences of (x, y) than points that have a color lower on the colorbar.

## Problem 3

Using a folding variable to make the histogram:

```python
folded = t%(2*np.pi)
plt.hist2d(folded, bkgd, 150)
plt.colorbar()
plt.show()
```

After folding our time over a period of 2$\pi$, we can see more clearly that the data is sinusoidal.
